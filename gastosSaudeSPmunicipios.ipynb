```python
import pandas as pd
```

```python
# mostra as cinco primeiras linhas da tabela de municípios
df = pd.read_csv('DespesasPrincipaisMunicipiosRegiaoSudeste.csv', sep=",", encoding="ANSI")
df.head()
```

```python
# Mostra valor gasto com saúde para cada município em 2020
df.groupby(['NO_ENTE'])['VALUE'].sum()
```

```python
# datas para cada período - municípios
df.groupby(['NO_ENTE'])['DT_STATUS_COLETA'].value_counts()
```
